```
def posicion_inicial():
    print("Quedarse parado en posición inicial")

def intercalar_piernas(n):
    for i in range(n):
        print("Intercalar pierna izquierda girando hacia un lado")
        print("Intercalar pierna derecha girando hacia un lado")

def patada(n):
    for i in range(n):
        print("Lanzar patada hacia un lado")

def caminata_con_brincos(n):
    for i in range(n):
        print("Caminata con brincos pequeños al ritmo de la música")

def multiples_patadas(n):
    for i in range(n):
        print("Lanzar múltiples patadas hacia un lado seguidas")

def mover_cabeza_pierna(n):
    for i in range(n):
        print("Mover cabeza hacia un lado mientras la pierna se mueve por detrás al ritmo de la música")

def saltos_piernas(n):
    for i in range(n):
        print("Abrir y cerrar piernas dando pequeños saltos hacia adelante con Manos agarradas a la cintura")


def baile_borde_interno(n):
    for i in range(n):
        print("Bailar con el borde interno de los pies al ritmo de la música")

def zombie(n):
    for i in range(n):
        print("Agacharse como zombie")
        print("Levantarse como zombie")

def remate():
    print("Pose final de Kung Fu 🥋")

def kung_fu_fighting(n):
    print("=" * 40)
    print("      🥋 KUNG FU FIGHTING 🥋")
    print("=" * 40)
    for i in range(n):
        print(f"\n--- Ronda {i+1} ---")
        posicion_inicial()
        intercalar_piernas(1)
        patada(1)
        caminata_con_brincos(1)
        multiples_patadas(1)
        mover_cabeza_pierna(1)
        caminata_con_brincos(1)
        saltos_piernas(3)
        patada(1)
        caminata_con_brincos(1)
        baile_borde_interno(1)
        zombie(2)

kung_fu_fighting(1)
remate()
```
