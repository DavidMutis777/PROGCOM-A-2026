```
class Calculadora:
    def __init__(self, capital, tasa, tiempo):
        self.capital = capital
        self.tasa = tasa / 100
        self.tiempo = tiempo

    def interes_simple(self):
        interes = self.capital * self.tasa * self.tiempo
        total = self.capital + interes
        return interes, total

    def interes_compuesto(self):
        total = self.capital * (1 + self.tasa) ** self.tiempo
        interes = total - self.capital
        return interes, total

def main():
    print("CALCULADORA DE INTERES SIMPLE Y COMPUESTO")
    capital = float(input("Capital inicial: "))
    tasa = float(input("Tasa de interes anual (%): "))
    tiempo = float(input("Tiempo (años): "))

    calc = Calculadora(capital, tasa, tiempo)

    interes_s, total_s = calc.interes_simple()
    interes_c, total_c = calc.interes_compuesto()

    print("\n--- INTERES SIMPLE ---")
    print(f"Interes generado: ${interes_s:.2f}")
    print(f"Total a pagar: ${total_s:.2f}")

    print("\n--- INTERES COMPUESTO ---")
    print(f"Interes generado: ${interes_c:.2f}")
    print(f"Total a pagar: ${total_c:.2f}")

main()
```
